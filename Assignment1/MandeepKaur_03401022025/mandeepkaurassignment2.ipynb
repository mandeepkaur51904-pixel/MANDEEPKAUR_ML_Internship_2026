```python
# program to find sum of first 10 natural numbers
total_sum = 0
for i in range(1, 11):
    total_sum += i  # Adds the current number to the running total
print("The sum of the first 10 natural numbers is:", total_sum)
```

```python
#program to find factorial of a number
num = int(input("Enter a number: "))
factorial = 1
if num < 0:
    print("Factorial does not exist for negative numbers.")
elif num == 0:
    print("The factorial of 0 is 1")
else:
    for i in range(1, num + 1):
        factorial = factorial * i
    print(f"The factorial of {num} is {factorial}")
```

```python
#program to print fibonacci series
n=int(input("enter a number"))
a, b = 0, 1
count = 0
if n <= 0:
  print("Please enter a positive integer.")
elif n == 1:
  print(f"Fibonacci sequence up to {n} term: {a}")
else:
  print("Fibonacci sequence:")
while count < n:
  print(a, end=" ")
  a, b = b, a + b
  count += 1
  print()
```

```python
# Student Result System
student_name = input("Enter Student Name: ")
roll_number = input("Enter Roll Number: ")
subjects = ["Mathematics", "Science", "English", "History", "Computer Science"]
marks_obtained = []
max_marks_per_subject = 100
print("\n--- Enter Marks (Out of 100) ---")
for subject in subjects:
    while True:
        try:
            mark = float(input(f"Enter marks for {subject}: "))
            if 0 <= mark <= max_marks_per_subject:
                marks_obtained.append(mark)
                break
            else:
                print("Invalid input! Marks must be between 0 and 100.")
        except ValueError:
            print("Invalid input! Please enter a valid number.")
total_marks = sum(marks_obtained)
max_total_marks = len(subjects) * max_marks_per_subject
percentage = (total_marks / max_total_marks) * 100
if percentage >= 90:
    grade = "A+"
elif percentage >= 80:
    grade = "A"
elif percentage >= 70:
    grade = "B"
elif percentage >= 60:
    grade = "C"
elif percentage >= 50:
    grade = "D"
else:
    grade = "Fail"
print("\n" + "="*40)
print("           STUDENT REPORT CARD          ")
print("="*40)
print(f"Student Name : {student_name}")
print(f"Roll Number  : {roll_number}")
print("-"*40)
for i in range(len(subjects)):
    print(f"{subjects[i]:<18} : {marks_obtained[i]} / {max_marks_per_subject}")

print("-"*40)
print(f"Total Marks  : {total_marks} / {max_total_marks}")
print(f"Percentage   : {percentage:.2f}%")
print(f"Final Grade  : {grade}")
print("="*40)
```
