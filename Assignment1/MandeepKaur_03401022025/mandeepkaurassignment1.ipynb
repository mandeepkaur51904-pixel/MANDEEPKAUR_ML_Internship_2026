```python
#square and cube of a number
num = float(input("Enter a number: "))
square = num ** 2
cube = num ** 3
print(f"The square of {num} is {square}")
print(f"The cube of {num} is {cube}")
```

```python
#swap two numbers without third variable
x=float(input("enter a number"))
y=float(input("enter another number"))
print("before swap",x,y)
x, y = y, x
print(f"After swap: x = {x}, y = {y}")
```

```python
# Python Program: Student Report Card Generator

# 1. Take student personal details using input()
student_name = input("Enter Student Name: ")
roll_number = input("Enter Roll Number: ")
student_class = input("Enter Class: ")
print("\n--- Enter Subject Marks (Out of 100) ---")
# 2. Store individual subject marks in separate variables
# float() converts the default string input to a decimal number
math_marks = float(input("Enter Marks for Mathematics: "))
science_marks = float(input("Enter Marks for Science: "))
english_marks = float(input("Enter Marks for English: "))
# 3. Calculate Total Marks
# Adds up the marks stored in the individual variables
total_marks = math_marks + science_marks + english_marks
# 4. Calculate Percentage
# Divide obtained total by maximum possible marks (3 subjects * 100 = 300) and multiply by 100
max_possible_marks = 300
percentage = (total_marks / max_possible_marks) * 100
# 5. Display the final Student Report Card
print("\n" + "="*40)
print("           STUDENT REPORT CARD          ")
print("="*40)
print(f"Student Name : {student_name}")
print(f"Roll Number  : {roll_number}")
print(f"Class        : {student_class}")
print("-"*40)
print(f"Mathematics  : {math_marks} / 100")
print(f"Science      : {science_marks} / 100")
print(f"English      : {english_marks} / 100")
print("-"*40)
# :.2f limits the decimal output to exactly 2 decimal places for clean formatting
print(f"Total Marks  : {total_marks} / {max_possible_marks}")
print(f"Percentage   : {percentage:.2f}%")
print("="*40)
```
