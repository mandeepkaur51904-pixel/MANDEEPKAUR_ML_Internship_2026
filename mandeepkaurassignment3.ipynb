```python
# create a function to print first 10 natural numbers
def print_ten_numbers():
    for i in range(1, 11):
        print(i)
print_ten_numbers()
```

```python
#create a function to calculate sum of first N natural numbers
def find_sum(n):
    return (n * (n + 1)) // 2
# Example usage
print(find_sum(5))  # Output: 15
```

```python
# create a function to reverse a number
def reverse_number_str(num):
    sign = -1 if num < 0 else 1
    reversed_num = int(str(abs(num))[::-1])
    return sign * reversed_num
print(reverse_number_str(1234))
print(reverse_number_str(-567))
print(reverse_number_str(1200))
```

```python
# create a function to count digits in a number
def count_digits_str(num):
    # abs handles negative numbers; str converts to text; len counts characters
    return len(str(abs(num)))

# Examples
print(count_digits_str(12345))
print(count_digits_str(-589))
print(count_digits_str(0))
```

```python
#create a function to check palindrome number
def is_palindrome(n):
    return str(n) == ''.join(reversed(str(n)))

# Get the number from the user
n = int(input("Enter number: "))

# Check if the number is a palindrome
if is_palindrome(n):
  print("The number is a palindrome!")
else:
  print("The number is not a palindrome.")
```

```python
#create a function to generate fibonacci series
def generate_fibonacci_list(n):
    # Handle base cases where elements are less than or equal to zero
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    # Start with the first two numbers of the sequence
    fib_series = [0, 1]
    # Loop to add the remaining numbers
    for _ in range(2, n):
        # Calculate next term by adding the last two elements
        next_term = fib_series[-1] + fib_series[-2]
        fib_series.append(next_term)
        return fib_series

```

```python
# Functions for basic arithmetic operations
def add(a, b): return a + b
def subtract(a, b): return a - b
def multiply(a, b): return a * b
def divide(a, b): return a / b if b != 0 else "Error: Division by zero"

def calculator():
    """Main calculator loop: User selects operation, inputs, gets result."""
    while True:
        print("\n1.Add 2.Sub 3.Mul 4.Div 5.Exit")
        choice = input("Select operation (1-5): ")
        if choice == '5': break

        try:
            n1 = float(input("First number: "))
            n2 = float(input("Second number: "))
            if choice == '1': print("Result:", add(n1, n2))
            elif choice == '2': print("Result:", subtract(n1, n2))
            elif choice == '3': print("Result:", multiply(n1, n2))
            elif choice == '4': print("Result:", divide(n1, n2))
        except ValueError:
            print("Invalid input!")

if __name__ == "__main__":
    calculator()
```

```python
# Program to create a file and save student details
filename = "students.txt"

# Collect data from the user
name = input("Enter student name: ")
roll_no = input("Enter roll number: ")
grade = input("Enter grade/marks: ")

# Open the file in append mode ('a') so it does not overwrite existing records
with open(filename, "a") as file:
    # Write details as a comma-separated line
    file.write(f"{name},{roll_no},{grade}\n")

print(f"Successfully saved details to {filename}!")
```

```python
# handle division by zero using exception handling
try:
    # Take two numbers as input
    num1 = int(input("Enter numerator: "))
    num2 = int(input("Enter denominator: "))

    # Perform division
    result = num1 / num2

    # Display result
    print("Result =", result)

# Handle division by zero error
except ZeroDivisionError:
    print("Error: Division by zero is not allowed.")

# Handle invalid input
except ValueError:
    print("Error: Please enter valid numbers.")

# Always executes
finally:
    print("Program executed successfully.")
```

```python
# Creating a Student class
class Student:

    # Constructor to initialize data members
    def _init_(self, name, marks):
        self.name = name
        self.marks = marks

    # Method to display student details
    def display(self):
        print("Student Name :", self.name)
        print("Student Marks:", self.marks)

# Taking input from user
name = input("Enter student name: ")
marks = int(input("Enter student marks: "))

# Creating object of Student class
student1 = Student(name, marks)

# Calling display method
student1.display()
```
